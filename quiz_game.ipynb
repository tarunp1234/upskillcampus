```python
import json
import time
import os
import random   

HIGH_SCORE_FILE = "highscore.txt"

# Load questions
def load_questions(file_path):
    with open(file_path, 'r',encoding='utf-8') as file:
        return json.load(file)["questions"]

# Load high score
def load_high_score():
    if not os.path.exists(HIGH_SCORE_FILE):
        return 0
    with open(HIGH_SCORE_FILE, "r") as file:
        return int(file.read())

# Save high score
def save_high_score(score):
    with open(HIGH_SCORE_FILE, "w") as file:
        file.write(str(score))

# Ask questions with timer
def ask_questions(questions):
    score = 0

    for i, q in enumerate(questions, start=1):
        print("\n" + "="*40)
        print(f"Q{i}: {q['question']}")

        for opt in q["options"]:
            print(opt)

        start_time = time.time()
        answer = input("Your answer (A/B/C/D): ").upper()
        end_time = time.time()

        # ⏱ Timer check
        if end_time - start_time > 10:
            print("⏰ Time's up!")
            continue

        # ⚠ Input validation
        if answer not in ['A', 'B', 'C', 'D']:
            print("⚠ Invalid input! Skipped.")
            continue

        # ✅ Answer check
        if answer == q["answer"]:
            print("✅ Correct!")
            score += 1
        else:
            print(f"❌ Wrong! Correct answer: {q['answer']}")

    return score

# Show result
def show_result(score, total):
    print("\n" + "="*40)
    print("📊 RESULT")
    print("="*40)

    percentage = (score / total) * 100
    print(f"Score: {score}/{total}")
    print(f"Percentage: {percentage:.2f}%")

    if percentage >= 80:
        print("🌟 Excellent!")
    elif percentage >= 50:
        print("👍 Good Job!")
    else:
        print("💪 Keep Practicing!")

# Filter by category
def filter_questions(questions, category):
    return [q for q in questions if q["category"].lower() == category.lower()]

# Main program
def main():
    print("\n🎯 WELCOME TO ADVANCED QUIZ GAME 🎯")

    questions = load_questions("questions.json")

    # 🔥 Shuffle all questions
    random.shuffle(questions)

    categories = list(set(q["category"] for q in questions))
    print("\nAvailable Categories:", ", ".join(categories))

    chosen = input("Choose category: ")

    filtered = filter_questions(questions, chosen)

    if not filtered:
        print("No questions found for this category.")
        return

    # 🔥 Shuffle filtered questions
    random.shuffle(filtered)

    # 🔥 Ask only 5 random questions (optional)
    filtered = random.sample(filtered, min(5, len(filtered)))

    high_score = load_high_score()
    print(f"\n🏆 Current High Score: {high_score}")

    score = ask_questions(filtered)

    show_result(score, len(filtered))

    if score > high_score:
        print("🎉 New High Score!")
        save_high_score(score)

    retry = input("\nDo you want to play again? (yes/no): ").lower()
    if retry == "yes":
        main()
    else:
        print("Thanks for playing! 👋")

if __name__ == "__main__":
    main()
```
